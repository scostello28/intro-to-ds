**Feature Engineering**
1. Create FG_percentage, 2P_percentage, 3P_percentage, FT_percentage and Points Columns

```python
# 1. Create FG_percentage, 2P_percentage, 3P_percentage, FT_percentage and Points Columns

# FG_perc
df['FG_perc'] = df['FG'] / df['FGA']

# 2P_perc
df['2P_perc'] = df['2P'] / df['2PA']

# 3P_perc
df['3P_perc'] = df['3P'] / df['3PA']

# FT_perc
df['FT_perc'] = df['FT'] / df['FTA']

# FT_perc
df['Points'] = df['FT'] +  df['2P'] * 2 + df['3P'] * 3
```

```python
df[['FG_perc', '2P_perc', '3P_perc', 'FT_perc', 'Points']].head(10)
```

**Query Questions**
1. How many teams are there?
2. Gonzaga Players by minutes played from 2018 Season
3. Senior Center with most turn overs, per 100 possessions, before 2017.
4. Team with most points, per 100 possessions, by under classmen.
5. Player under 6 feet tall with most fouls--per 100 possessions.
6. Guards from each season in top 1% for three-point shot percentage, per 100 possessions, from the 2018 season.
7. How many rows have null values?
8. Mean of Games played by players with null values.
9. Mean of Games played by players without null values.

```python
# 1. How many teams are there?
df.Team.nunique()
len(df.Team.unique())
```

```python
# 2. Gonzaga Players sorted by minutes played from 2018 Season
df[(df['Season'] == 2018) & (df['Team'] == 'gonzaga')].sort_values('MP', ascending=False)
```

```python
# 3. Senior Center with most turn overs, per 100 possessions, before 2017.
cond1 = (df['Class'] == 4)
cond2 = (df['Pos'] == 'C')
cond3 = (df['Season'] < 2017)
no2_df = df[cond1 & cond2 & cond3]  #.reset_index(drop=True)
# no2_df
no2_df = no2_df[['Team', 'Player', 'TOV']].groupby('Player').sum()  #.reset_index()
# no2_df
no2_df.sort_values('TOV', ascending=True).iloc[-1]
```

```python
# 4. Team with most points, per 100 possessions, by under classmen since 2016.
no3_df = df[(df['Class'] < 3) & (df['Season'] >= 2016)]
no3_df = no3_df.groupby('Team').sum()
# no3_df
no3_df = no3_df[['Points', '2P', '3P']]
# no3_df = no3_df['Points']
# no3_df
no3_df[no3_df.Points == no3_df.Points.max()]
```

```python
# 5. 5 players under 6 feet tall with most fouls--per 100 possessions
no4_df = df[df['Height'] < 72]
no4_df = no4_df.groupby('Player').sum()['PF']
no4_df.sort_values(ascending=False).iloc[:5]
```

```python
df.dtypes
```

```python
# 6. Guards in top 1% for three-point shot percentage, per 100 possessions, from the 2018 season.
no5_df = df[(df['Season'] == 2018) & (df['Pos'] == 'G')] #[['Player','3P_perc', 'MP']]
no5_df = no5_df[['Team', 'Pos', 'G', 'MP','Player', '3P', '3PA','3P_perc']]
# no5_df
no5_df = no5_df[no5_df['MP'] > 250]
no5_df = no5_df[no5_df['3P_perc'] >= no5_df['3P_perc'].quantile(.99)]
no5_df.sort_values('3P_perc', ascending=False).reset_index(drop=True)
```
