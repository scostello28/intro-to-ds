## Functions

```python
# np.<tab> to see all numpy functions
```

```python
j = np.array([-2, 5, -4, 6])
```

```python
print("Sum: {}".format(np.sum(j)))
print("Prduct: {}".format(np.prod(j)))
print("Mean: {}".format(np.mean(j)))
print("Standard Dev: {:.4f}".format(np.std(j)))
print("Variance: {:.4f}".format(np.var(j)))
print("Min: {}".format(np.min(j)))
print("Max: {}".format(np.max(j)))
print("Min index: {}".format(np.argmin(j))) 
print("Max index: {}".format(np.argmax(j)))
print("absolute value: {}".format(np.abs(j)))
print("sine: {}".format(np.sin(j)))
```

```python
print("Number of Dimensions: {}".format(j.ndim))           # The number of axes (dimensions) of the array. Number of dimensions is referred to as rank.
print("Shape: {}".format(j.shape))          # The dimensions of the array. Tuple of integers indicating the size of the array in each dimension. Shape (n,m) -> (rows, cols) -> n = rows, m = cols. The length of shape tuple is the rank, or number of dimensions, ndim.
print("Size: {}".format(j.size))           # The total number of elements of the array. This is equal to the product of the elements of shape.
print("Data Types: {}".format(j.dtype))          # Object describing the type of the elements in the array. One can create or specify dtypes using standard Python types. Additionally NumPy provides types of its own. Ex. numpy.int32, numpy.int16, and numpy.float64.
print("Item Size: {}".format(j.itemsize))       # Size in bytes of each element of the array. For example, an array of elements of type float64 has itemsize 8 (=64/8), while one of type complex32 has itemsize 4 (=32/8). It is equivalent to ndarray.dtype.itemsize.
print("Data: {}".format(j.data))           # Buffer containing the actual elements of the array. Normally, we wont need to use this attribute because we will access the elements in an array using indexing facilities.
```

## Vector and Matrix Operations

```python
# Elementwise operations of vectors
y = np.array([1,2,3])
x = np.array([2,3,4])

y_add_x = y + x
y_subtract_x = y - x
y_divide_x = y / x

# Also called Hadamard Product (when it outputs a vector, not dot product)
y_multiply_x = y * x
```

```python
print(y_add_x)
print(y_subtract_x)
print(y_divide_x)
print(y_multiply_x)
```

```python
# Dot product of vectors
y = np.array([1,2,3])
x = np.array([2,3,4])
y_dot_x = np.dot(y,x)
```

```python
print(y_dot_x)
```

```python
# Make matrix in NumPy
a = np.array([
 [1,2,3],
 [4,5,6]
])
b = np.array([
 [1,2,3]
])
```

```python
a
```

```python
# Check the dimensions of your matrix
print(a.shape)
print(b.shape)
```

```python
# Scalar Operations with Matrices
a_add_1 = a + 1
a_sub_1 = a - 1
a_divide_1 = a / 1
a_multiply_1 = a * 1
```

```python
print(a_add_1)
print(a_sub_1)
print(a_divide_1)
print(a_multiply_1)
```

```python
# Elementwise Operations with Matrices
a = np.array([
 [1,2],
 [3,4]])
b = np.array([
 [1,2],
 [3,4]])

a_add_b = a + b
a_sub_b = a - b

# Elementwise product of matrices (not dot product)
a_multiply_b = a * b
```

```python
print(a_add_b)
print(a_sub_b)
print(a_multiply_b)
```

```python
# Matrix Transposition
a = np.array([
   [1, 2],
   [3, 4]])

print(a)
print(a.T)
```

```python
# Matrix Multiplication 
# Remember: num columns from 1st matrix must == num rows from 2nd
# Remember: resulting matrix will have shape (rows of 1st, cols of 2nd)
a = np.array([
 [1, 2]
 ])
b = np.array([
 [3, 4],
 [5, 6]
 ])

# Multiply
mm = np.dot(a,b)
```

```python
print(a.shape)
print(b.shape)
print(mm.shape)
```

```python
# What happens if you try to dot product matrices with the wrong shape?
a = np.array([
 [1, 2]
 ])
b_wrong_dim = np.array([
 [3, 4],
 [5, 6],
    [7,5]
 ])
```

```python
print(a.shape)
print(b_wrong_dim.shape)
```

```python
# Fail at dot product
# You will get this error more than any other error. I promise.
mm_fail = np.dot(a,b_wrong_dim)
```

```python
# But, you can still elementwise multiply them, which can cause some bugs that are hard to find.
mm_hadamard = a * b_wrong_dim
print(mm_hadamard)
```

```python
# However, if you have a matrix is the wrong dim, you can transpose it to fit the other matrix if you can get the cols from first to match rows from second
print(a.shape)
print(b_wrong_dim.shape)
```

```python
mm_dot_bT = np.dot(a,b_wrong_dim.T)
```

```python
print(mm_dot_bT)
```
